A interface é dividida em:

- **Células de Código**: Estas células executam código Python, se você não é programador basta simplesmente ignorar o código e clicar no botão play (apontamento em vermelho).

A saída é o que realmente nos interessa, é o resultado da instrução fornecida, é aqui que você deve se focar.

Para executar uma célula de código, use o atalho `Shift + Enter` ou clique no ícone de "Play" à esquerda da célula. Este é um código divertido que vai gerar um sugestão para seu almoço.

### Exemplo: Executando um Código em Python



```python
import random

# Lista de opções de comidas com emojis
comidas = ["🍕 Pizza", "🍔 Hambúrguer", "🥗 Salada", "🍣 Sushi", "🍜 Lámen", "🍝 Macarrão", "🥪 Sanduíche", "🌮 Taco", "🍛 Curry", "🥘 Paella", "🍤 Camarão", "🍲 Sopa"]

# Função para sortear a sugestão de almoço
def sugestao_almoco():
    return random.choice(comidas)

# Mostra a sugestão de almoço
print("🔄 Sugestão de Almoço 🔄")
print("Hoje você pode experimentar:", sugestao_almoco())
```
